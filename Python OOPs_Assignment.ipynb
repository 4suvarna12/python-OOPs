8.How is encapsulation achieved in Python?
- Encapsulation in Python is achieved through a combination of naming conventions and access modifiers
- Wrapping data and funtions into a single class.
- Here's how encapsulation is typically implemented in Python:

1. Naming Conventions (The Primary Mechanism):

- Public Members: Attributes and methods that are intended to be accessed from outside the class are usually named with a standard, non-prefixed identifier (e.g., name, age, calculate_area()). This signals that these members are part of the object's public interface.

- "Protected" Members (Single Leading Underscore): Attributes and methods prefixed with a single underscore (_) are conventionally treated as "protected." This suggests that they are intended for internal use within the class and its subclasses. While Python doesn't strictly prevent external access to these members, it's a strong convention that programmers should respect. Modifying these directly from outside is generally discouraged as it can break the class's internal logic.

- "Private" Members (Double Leading Underscore - Name Mangling): Attributes and methods prefixed with a double underscore (__) have a special behavior called name mangling. Python renames these members internally to make it harder (but not impossible) to access them directly from outside the class. The name is changed to _ClassName__membername. This provides a stronger form of "privacy" compared to the single underscore convention

4. Abstraction:

- Abstraction involves hiding complex implementation details and showing only the essential information to the user.
- OOP achieves abstraction through classes and objects, focusing on what an object does rather than how it does it.
- This simplifies the user's interaction with objects and makes the code easier to understand and work with at a higher level.

17.What is the difference between a class variable and an instancevariable?
- Class Variable:

- Definition: A class variable is defined within the class but outside of any instance methods (like __init__, or other regular methods).
- Scope: Class variables belong to the class itself. They are shared among all instances (objects) of that class.
- Access:They can be accessed directly using the class name (e.g., ClassName.class_variable).

- instance variable:
- Definition: An instance variable is defined inside instance methods, usually within the __init__ method using the self keyword (e.g., self.instance_variable = value).
- Scope: Instance variables belong to the specific instance (object) of the class. Each instance has its own independent copy of the instance variables.
- Access: Instance variables are accessed using the instance object followed by the dot notation (e.g., instance.instance_variable). You cannot directly access them using the class name alone.

```python
# 2.Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.

from abc import ABC,abstractmethod

class Shape(ABC):
  @abstractmethod
  def area(self):
    pass

class Circel(Shape):
  def area(self,radius):
    return 3.14*radius*radius

class Rectangle(Shape):
  def area(self,length,width):
    return length*width

#Example
c1=Circel()
print(c1.area(10))

r1=Rectangle()
print(r1.area(10,20))
```

```python
# 5 Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance.

class BankAccount:
  def __init__(self,balance):
    self.__Balance = balance
    print(f"Initial balance is {self.__Balance}")
  def deposit(self,amount):
    self.__Balance += amount
    print(f"After deposit rs. {amount} balance is {self.__Balance}")

  def withdraw(self,amount):
    if amount <= self.__Balance:
      self.__Balance -= amount
      print(f"After withdraw rs. {amount} balance is {self.__Balance}")
    else:
      print("Insufficient balance")
  def get_balance(self):
    return f"Total balance {self.__Balance}"
  def set_balance(self,balance):
    self.__Balance = balance


#Example
account = BankAccount(1000)
account.deposit(500)
account.withdraw(200)
print(account.get_balance())
```

9.Write a class Fraction with attributes numerator and denominator. Override the str method to display the
fraction as "numerator/denominator".

```python
class Fraction:

    def __init__(self, numerator, denominator):

        if denominator == 0:
            raise ValueError("Denominator cannot be zero.")
        self.numerator = numerator
        self.denominator = denominator

    def __str__(self):

        return f"{self.numerator}/{self.denominator}"


    def __repr__(self):

        return f"Fraction(numerator={self.numerator}, denominator={self.denominator})"

if __name__ == "__main__":
    # Example Usage
    frac1 = Fraction(2, 3)
    frac2 = Fraction(5, 7)

    # Using str() implicitly (inside print)
    print(f"Fraction 1: {frac1}")  # Output: Fraction 1: 2/3
    print(f"Fraction 2: {frac2}")  # Output: Fraction 2: 5/7

    # Using str() explicitly
    print(f"String representation of frac1: {str(frac1)}") #Output: String representation of frac1: 2/3

    # Using repr()
    print(f"Representation of frac1: {repr(frac1)}")  # Output: Representation of frac1: Fraction(numerator=2, denominator=3)

    # Example of error handling
    try:
        frac3 = Fraction(1, 0)
    except ValueError as e:
        print(f"Error creating fraction: {e}")  # Output: Error creating fraction: Denominator cannot be zero.
```

10.Demonstrate operator overloading by creating a class Vector and overriding the add method to add two
vectors.

```python
class Vector:

    def __init__(self, coordinates):

        if not isinstance(coordinates, (list, tuple)):
            raise TypeError("Coordinates must be a list or tuple.")
        if not coordinates:
            raise ValueError("Coordinates cannot be empty.")
        if not all(isinstance(coord, (int, float)) for coord in coordinates):
            raise TypeError("All coordinates must be numbers.")

        self.coordinates = tuple(coordinates)  # Store as tuple for immutability
        self.dimension = len(coordinates)

    def __add__(self, other):

        if not isinstance(other, Vector):
            raise TypeError("Cannot add Vector with an object of different type.")
        if self.dimension != other.dimension:
            raise ValueError("Vectors must have the same dimension to be added.")

        new_coordinates = [self.coordinates[i] + other.coordinates[i] for i in range(self.dimension)]
        return Vector(new_coordinates)

    def __str__(self):

        return f"Vector({', '.join(map(str, self.coordinates))})"

    def __repr__(self):

        return f"Vector(coordinates={self.coordinates})"

    def __eq__(self, other):
        """
        Overrides the '==' operator to compare two vectors for equality.
        """
        if not isinstance(other, Vector):
            return False  # Not equal if 'other' is not a Vector
        return self.coordinates == other.coordinates

if __name__ == "__main__":
    # Example Usage
    v1 = Vector([1, 2, 3])
    v2 = Vector([4, 5, 6])
    v3 = Vector([1, 2, 3])  # For testing equality

    print(f"Vector 1: {v1}")
    print(f"Vector 2: {v2}")

    # Vector addition using the overloaded '+' operator
    v_sum = v1 + v2
    print(f"Sum of v1 and v2: {v_sum}")  # Output: Vector(5, 7, 9)

    #Demonstrates the __eq__
    print(f"v1 == v2: {v1 == v2}") #Output: v1 == v2: False
    print(f"v1 == v3: {v1 == v3}") #Output: v1 == v3: True

    try:
        v4 = Vector([1, 2])
        v5 = Vector([1, 2, 3])
        v6 = v4 + v5  # This will raise a ValueError
    except ValueError as e:
        print(f"Error: {e}")

    try:
        v7 = v1 + [1,2,3] #This will raise a TypeError
    except TypeError as e:
        print(f"Error: {e}")
```

```python
class Employee:
    """
    Represents an employee with a basic salary calculation based on hours worked and hourly rate.
    """
    def __init__(self, name, hourly_rate, hours_worked):
        """
        Initializes an Employee object.

        Args:
            name (str): The name of the employee.
            hourly_rate (float): The hourly rate of pay.
            hours_worked (float): The number of hours worked.

        Raises:
            TypeError: If name is not a string, or if hourly_rate or hours_worked are not numbers.
            ValueError: If hourly_rate or hours_worked are negative.
        """
        if not isinstance(name, str):
            raise TypeError("Name must be a string.")
        if not isinstance(hourly_rate, (int, float)):
            raise TypeError("Hourly rate must be a number.")
        if not isinstance(hours_worked, (int, float)):
            raise TypeError("Hours worked must be a number.")
        if hourly_rate < 0:
            raise ValueError("Hourly rate cannot be negative.")
        if hours_worked < 0:
            raise ValueError("Hours worked cannot be negative.")

        self.name = name
        self.hourly_rate = hourly_rate
        self.hours_worked = hours_worked

    def calculate_salary(self):
        """
        Calculates the salary of the employee.

        Returns:
            float: The calculated salary (hours_worked * hourly_rate).
        """
        return self.hours_worked * self.hourly_rate

    def __str__(self):
        """
        Returns a string representation of the Employee object.
        """
        return f"Employee(name='{self.name}', hourly_rate={self.hourly_rate}, hours_worked={self.hours_worked})"

    def __repr__(self):
        """
        Official string representation for developers.
        """
        return f"Employee(name='{self.name}', hourly_rate={self.hourly_rate}, hours_worked={self.hours_worked})"



class Manager(Employee):
    """
    Represents a manager, who is a type of employee with a bonus added to their salary.
    """
    def __init__(self, name, hourly_rate, hours_worked, bonus):
        """
        Initializes a Manager object.

        Args:
            name (str): The name of the manager.
            hourly_rate (float): The hourly rate of pay.
            hours_worked (float): The number of hours worked.
            bonus (float): The bonus to be added to the salary.

        Raises:
            TypeError: If bonus is not a number.
            ValueError: If bonus is negative.
        """
        super().__init__(name, hourly_rate, hours_worked)  # Call the Employee constructor
        if not isinstance(bonus, (int, float)):
            raise TypeError("Bonus must be a number.")
        if bonus < 0:
            raise ValueError("Bonus cannot be negative.")
        self.bonus = bonus

    def calculate_salary(self):
        """
        Calculates the salary of the manager, including the bonus.
        Overrides the calculate_salary() method of the Employee class.

        Returns:
            float: The calculated salary (hours_worked * hourly_rate + bonus).
        """
        # Call the Employee class's calculate_salary() and add bonus
        return super().calculate_salary() + self.bonus

    def __str__(self):
        """
        Returns a string representation of the Manager object.
        """
        return f"Manager(name='{self.name}', hourly_rate={self.hourly_rate}, hours_worked={self.hours_worked}, bonus={self.bonus})"

    def __repr__(self):
        """
        Official string representation for developers.
        """
        return f"Manager(name='{self.name}', hourly_rate={self.hourly_rate}, hours_worked={self.hours_worked}, bonus={self.bonus})"

if __name__ == "__main__":
    # Example Usage
    employee1 = Employee("Alice", 20, 40)
    print(employee1)
    print(f"{employee1.name}'s salary: ${employee1.calculate_salary():.2f}")  # Output: Alice's salary: $800.00

    manager1 = Manager("Bob", 30, 40, 500)
    print(manager1)
    print(f"{manager1.name}'s salary: ${manager1.calculate_salary():.2f}")  # Output: Bob's salary: $1700.00

    #Demonstrate that a Manager is also an Employee.
    print(isinstance(manager1, Employee)) #True
    print(isinstance(employee1, Manager)) #False

    try:
        employee2 = Employee("Eve", -10, 40)  # Raises ValueError
    except ValueError as e:
        print(f"Error creating employee: {e}")

    try:
        manager2 = Manager("Frank", 25, 30, -100)  # Raises ValueError
    except ValueError as e:
        print(f"Error creating manager: {e}")

    try:
        manager3 = Manager("Grace", 25, 30, "invalid")  # Raises TypeError
    except TypeError as e:
        print(f"Error creating manager: {e}")
```

18.Create a class House with attributes address and price. Create a derived class Mansion that adds an
attribute number_of_rooms.

```python
class House:
    """
    Represents a house with an address and a price.
    """
    def __init__(self, address, price):
        """
        Initializes a House object.

        Args:
            address (str): The address of the house.
            price (int or float): The price of the house.
        """
        # Input validation: Check data types and values
        if not isinstance(address, str):
            raise TypeError("Address must be a string")
        if not isinstance(price, (int, float)):
            raise TypeError("Price must be a number (int or float)")
        if price < 0:
            raise ValueError("Price must be a non-negative number")

        self.address = address
        self.price = price

    def __str__(self):
        """
        Returns a user-friendly string representation of the House.
        """
        return f"House at {self.address} with price ${self.price}"

    def __repr__(self):
        """
        Returns a developer-friendly string representation of the House.
        """
        return f"House(address='{self.address}', price={self.price})"


class Mansion(House):
    """
    Represents a mansion, which is a type of house with an additional
    attribute: the number of rooms.
    """
    def __init__(self, address, price, number_of_rooms):
        """
        Initializes a Mansion object.

        Args:
            address (str): The address of the mansion.
            price (int or float): The price of the mansion.
            number_of_rooms (int): The number of rooms in the mansion.
        """
        # Call the parent class's constructor to handle address and price.
        super().__init__(address, price)

        # Input validation: Check the data type and value for number_of_rooms
        if not isinstance(number_of_rooms, int):
            raise TypeError("Number of rooms must be an integer")
        if number_of_rooms <= 0:
            raise ValueError("Number of rooms must be a positive integer")

        self.number_of_rooms = number_of_rooms

    def __str__(self):
        """
        Returns a user-friendly string representation of the Mansion.
        """
        return f"Mansion at {self.address} with {self.number_of_rooms} rooms and price ${self.price}"

    def __repr__(self):
        """
        Returns a developer-friendly string representation of the Mansion.
        """
        return f"Mansion(address='{self.address}', price={self.price}, number_of_rooms={self.number_of_rooms})"


if __name__ == "__main__":
    # Example Usage
    house1 = House("123 Main St", 250000)
    print(house1)  # Output: House at 123 Main St with price $250000
    print(repr(house1)) # Output: House(address='123 Main St', price=250000)

    mansion1 = Mansion("456 Luxury Ln", 1000000, 15)
    print(mansion1)  # Output: Mansion at 456 Luxury Ln with 15 rooms and price $1000000
    print(repr(mansion1)) # Output: Mansion(address='456 Luxury Ln', price=1000000, number_of_rooms=15)

    # Example of using the objects
    print(f"The price of {house1.address} is ${house1.price}")
    print(f"The mansion at {mansion1.address} has {mansion1.number_of_rooms} rooms.")

    #Illustrate that a Mansion is a House
    print(isinstance(mansion1, House)) #True
    print(isinstance(house1, Mansion)) #False

    # Example of error handling
    try:
        house2 = House("Invalid Address", -100)  # Invalid price
    except ValueError as e:
        print(f"Error creating house: {e}")

    try:
        mansion2 = Mansion("Fancy Place", 2000000, 0)  # Invalid number of rooms
    except ValueError as e:
        print(f"Error creating mansion: {e}")

    try:
        mansion3 = Mansion(123, 2000000, 10)  # Invalid address type
    except TypeError as e:
        print(f"Error creating mansion: {e}")
```
